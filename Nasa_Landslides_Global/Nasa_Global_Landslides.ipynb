# Temporal Analysis:

1. How has the frequency of rainfall-triggered landslide events varied over time since 2007?

```python
df.loc[:, 'event_date'] = pd.to_datetime(df['event_date'], format='%m/%d/%Y %I:%M:%S %p')
df = df[df['event_date'].dt.year >= 2007].copy()  
df.loc[:, 'year'] = df['event_date'].dt.year
events_per_year = df.groupby('year').size()
plt.figure(figsize=(10, 6))
events_per_year.plot(kind='line', marker='o', color='teal')
plt.title('Frequency of Rainfall-Triggered Landslide Events (2007-2016)')
plt.xlabel('Year')
plt.ylabel('Number of Events')
plt.grid(True)
plt.show()
```

2. Are there any seasonal trends in landslide occurrences?

```python
df['event_date'] = pd.to_datetime(df['event_date'], format='%m/%d/%Y %I:%M:%S %p')
df['month'] = df['event_date'].dt.month
events_per_month = df.groupby('month').size()
colors = ['blue', 'green', 'cyan', 'orange', 'purple', 'yellow', 'red', 'magenta', 'brown', 'pink', 'gray', 'lightblue']
plt.figure(figsize=(10, 6))
events_per_month.plot(kind='bar', color=colors)
plt.title('Seasonal Trends in Landslide Occurrences')
plt.xlabel('Month')
plt.ylabel('Number of Events')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```

# Spatial Analysis:

1. Which regions of the world have experienced the highest number of landslide events?

```python
events_per_region = df['location_description'].value_counts()
top_regions = events_per_region.head(10)  
plt.figure(figsize=(10, 6))
top_regions.plot(kind='bar', color='coral')
plt.title('Regions with Highest Frequency of Landslide Events')
plt.xlabel('Region')
plt.ylabel('Number of Events')
plt.xticks(rotation=45, ha='right')
plt.show()
```

2. Identify any geographical patterns or hotspots in landslide occurrences?

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='longitude', y='latitude', alpha=1)
plt.title('Geographical Distribution of Landslide Occurrences')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```

3. Have there been any significant impacts reported as a result of these landslides?

```python
impact_attributes = ['fatality_count', 'injury_count']

if all(attribute in df.columns for attribute in impact_attributes):
    summary_stats = df[impact_attributes].sum()
    plt.figure(figsize=(10, 6))
    summary_stats.plot(kind='bar', color='indigo', alpha=0.7)
    plt.title('Summary of Landslide Impacts')
    plt.xlabel('Impact Type')
    plt.ylabel('Total Count')
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.show()
    print("Summary of Landslide Impacts:")
    print(summary_stats)
else:
    print("One or more impact attributes are not available in the dataset.")
```

# Longitudinal Analysis:
1. How has the reporting of landslide events evolved over time?

```python
df['event_date'] = pd.to_datetime(df['event_date'])
df['year'] = df['event_date'].dt.year
events_per_year = df.groupby('year').size()
plt.figure(figsize=(10, 6))
events_per_year.plot(kind='line', color='slateblue', marker='o')
plt.title('Landslide Events Reported Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Events')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

2. Are there any changes in reporting accuracy or data quality over the years?

```python
df['year'] = pd.to_datetime(df['event_date']).dt.year
missing_values_percentage = df.groupby('year').apply(lambda x: x.isnull().mean()) * 100
plt.figure(figsize=(10, 6))
missing_values_percentage.plot(kind='line', marker='o')
plt.title('Percentage of Missing Values Over Time')
plt.xlabel('Year')
plt.ylabel('Percentage of Missing Values')
plt.grid(True)
plt.show()
```

The above visualization shows the percentage of missing values over time. The 'event_date' column is converted to datetime format, and the year is extracted. Then, the percentage of missing values for each year is calculated and plotted as a line graph.
